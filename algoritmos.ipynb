```python
class BinarySearch():

 def search_iterative(self, list, item):
    # Baixo e alto acompanham a parte de lista que voce esta procurando
    low = 0
    high = len(list) - 1

    # Enquanto ainda nao conseguiu chegar a um unico elemento
    while low <= high:
      # ... verificar o elemento central
      mid = (low + high) // 2
      guess = list[mid]
      # Acha o item
      if guess == item:
        return mid
      # O chute foi muito alto 
      if guess > item:
        high = mid - 1
      # O chute foi muito baixo
      else:
        low = mid + 1

    # O item nao existe
    return None

 def search_recursive(self, list, low, high, item):
    # Checar o caso base
    if high >= low:
        
        mid = (high + low) // 2
        guess = list[mid]
        
        # Se o elemento e presente no meio dele
        if guess == item:
            return mid
        
        # Se o elemento e menor que o meio, entao ele somente pode
        # Se presente no lado esquerdo do subarray
        elif guess > item:
              return self.search_recursive(list, low, mid - 1, item)
        
        # Falso elemento pode somente ser presente no lado direito do subarray
        else:
            return self.search_recursive(list, mid + 1, high, item)
        
    else:
        # Elemento nao esta presente no array
        return None

if __name__ == "__main__":
  # We must initialize the class to use the methods of this class
  bs = BinarySearch()
  my_list = [1, 3, 5, 7, 9]
  
  print(bs.search_iterative(my_list, 3)) # => 1

  # 'None' significa nulo em Python. Ele indica que o item nao foi encontrado 
  print(bs.search_iterative(my_list, -1)) # => None
```
